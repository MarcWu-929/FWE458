# Plot

```python
# MODIS data
modis_2013["Year"] = 2013
modis_2014["Year"] = 2014
modis_2015["Year"] = 2015
modis_2016["Year"] = 2016
modis_2017["Year"] = 2017
modis_2018["Year"] = 2018
modis_2019["Year"] = 2019
modis_2020["Year"] = 2020
modis_2021["Year"] = 2021
modis_2022["Year"] = 2022
modis_2023["Year"] = 2023

# CPR data
cpr_2013["Year"] = 2013
cpr_2014["Year"] = 2014
cpr_2015["Year"] = 2015
cpr_2016["Year"] = 2016
cpr_2017["Year"] = 2017
cpr_2018["Year"] = 2018
cpr_2019["Year"] = 2019
cpr_2020["Year"] = 2020
cpr_2021["Year"] = 2021
cpr_2022["Year"] = 2022
cpr_2023["Year"] = 2023

# Combine MODIS dataframes into one
modis_combined = pd.concat([modis_2013, modis_2014, modis_2015, modis_2016, modis_2017, modis_2018, modis_2019, modis_2020, modis_2021, modis_2022, modis_2023], ignore_index=True)

# Combine CPR dataframes into one
cpr_combined = pd.concat([cpr_2013, cpr_2014, cpr_2015, cpr_2016, cpr_2017, cpr_2018, cpr_2019, cpr_2020, cpr_2021, cpr_2022, cpr_2023], ignore_index=True)

# Define a list of states to include (12)
selected_states = ['Illinois', 'Indiana', 'Iowa', 'Kansas', 'Michigan', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota', 'Ohio', 'South Dakota', 'Wisconsin']
# all states (18)
# selected_states = ['Illinois', 'Indiana', 'Iowa', 'Kansas', 'Michigan', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota', 'Ohio', 'South Dakota', 'Wisconsin', 'Colorado', 'North Carolina', 'Pennsylvania', 'Tennessee', 'Texas']

# Filter the DataFrames for only the selected states
modis_filtered = modis_combined[modis_combined["NAME"].isin(selected_states)]
cpr_filtered = cpr_combined[cpr_combined["State"].isin(selected_states)]

# Set up the plot
fig, axs = plt.subplots(3, 4, figsize=(14, 18))

# List of MODIS phenological stages and colors
modis_stages = ["Greenup_1", "MidGreenup_1", "Maturity_1", "Peak_1",
                "Senescence_1", "MidGreendown_1", "Dormancy_1"]
modis_colors = ['green', 'limegreen', 'orange', 'red', 'brown', 'purple', 'gray']

# List of CPR phenological stages and colors
cpr_stages = ["Planted", "Emerged", "Silking", "Dough", "Dented", "Mature", "Harvested"]
cpr_colors = ['blue', 'cyan', 'magenta', 'olive', 'black', 'pink', 'teal']

# Initialize empty lists for legend handles and labels
modis_handles = []
modis_labels = []
cpr_handles = []
cpr_labels = []

# Loop through each axis and plot for each state
for i, ax in enumerate(axs.flat):
    if i < len(modis_filtered["NAME"].unique()):
        state = modis_filtered["NAME"].unique()[i]
        state_modis_data = modis_filtered[modis_filtered["NAME"] == state]
        state_cpr_data = cpr_filtered[cpr_filtered["State"] == state]
        ax.set_title(state)

        # Plot MODIS data
        for stage, color in zip(modis_stages, modis_colors):
            # Plot MODIS points for each year
            for year in range(2013, 2024):
                ax.plot(state_modis_data[state_modis_data["Year"] == year]["Year"],
                        state_modis_data[state_modis_data["Year"] == year][stage],
                        color=color, marker='o', linestyle='-', markersize=5)

            # Connect MODIS points for the same phenological stage with lines
            ax.plot(range(2013, 2024), [state_modis_data[state_modis_data["Year"] == year][stage].values[0] for year in range(2013, 2024)],
                    color=color, linestyle="-", linewidth=1)

            # Add MODIS legend handles only once for each stage
            if stage not in modis_labels:
                modis_labels.append(stage)
                modis_handles.append(ax.plot([], [], color=color, marker='o', linestyle='-', markersize=5)[0])

        # Plot CPR data for each year (2013-2023) with lines connecting data points
        for stage, color in zip(cpr_stages, cpr_colors):
            # Sort CPR data by year
            sorted_cpr_data = state_cpr_data.sort_values(by="Year")

            # Connect CPR points with lines
            ax.plot(sorted_cpr_data["Year"], sorted_cpr_data[stage],
                    color=color, marker='x', linestyle='--', markersize=6)

            # Add CPR legend handles only once for each stage
            if stage not in cpr_labels:
                cpr_labels.append(stage)
                cpr_handles.append(ax.plot([], [], color=color, marker='^', linestyle='-', markersize=6)[0])

        ax.set_xticks(range(2013, 2024, 2))  # Set x-axis to show every other year
        ax.set_xticklabels(range(2013, 2024, 2))  # Label x-axis
        ax.set_yticks(range(50, 350, 50))  # Set y-axis to show DOY (0 to 365)
        ax.set_yticklabels(range(50, 350, 50))  # Label y-axis as DOY

# Add global legend at the bottom of the figure
fig.legend(modis_handles + cpr_handles, modis_labels + cpr_labels, title='Phenological Stages', loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=4)
fig.supxlabel("Year", fontsize=20, fontweight="bold")  # Global x-axis label
fig.supylabel("DOY", fontsize=20, fontweight="bold")   # Global y-axis label

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
