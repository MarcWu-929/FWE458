```python
# List of years
years = list(range(2013, 2024))

# Load all dataframes into a dictionary with year as the key
modis_dfs = {
    year: pd.read_csv(f'/content/drive/MyDrive/Lab/CPR/2013/MODIS/MODIS_States_{year}.csv')[
        ['NAME', 'Greenup_1', 'MidGreenup_1', 'Maturity_1', 'Peak_1',
         'Senescence_1', 'MidGreendown_1', 'Dormancy_1']
    ].sort_values(by=['NAME'])
    for year in years
}

# Rename columns with suffixes
for year, df in modis_dfs.items():
    df.rename(columns={col: f"{col}_{year}" for col in df.columns if col != 'NAME'}, inplace=True)

# Merge all dataframes on 'NAME'
merged_df = reduce(lambda left, right: pd.merge(left, right, on='NAME', how='outer'), modis_dfs.values())

# Display merged dataframe
merged_df.head()
```

```python
# List of phenology stages (excluding 'NAME')
phenology_stages = ['Greenup_1', 'MidGreenup_1', 'Maturity_1', 'Peak_1',
                    'Senescence_1', 'MidGreendown_1', 'Dormancy_1']

# Create empty dictionaries to store mean and std values
mean_df = pd.DataFrame({'NAME': merged_df['NAME']})  # Retain state names
std_df = pd.DataFrame({'NAME': merged_df['NAME']})   # Retain state names

# Compute mean and standard deviation for each phenology stage
for stage in phenology_stages:
    stage_columns = [col for col in merged_df.columns if col.startswith(stage)]

    # Calculate mean and std across years, rounding to integers
    mean_df[f'{stage}_mean'] = merged_df[stage_columns].mean(axis=1, skipna=True).round(0).astype(int)
    std_df[f'{stage}_std'] = merged_df[stage_columns].std(axis=1, skipna=True).round(0).astype(int)

# Merge mean and std dataframes
stats_df = pd.merge(mean_df, std_df, on='NAME')

# List of states to keep
selected_states = ['Illinois', 'Indiana', 'Iowa', 'Kansas', 'Michigan', 'Minnesota',
                   'Missouri', 'Nebraska', 'North Dakota', 'Ohio', 'South Dakota', 'Wisconsin']

# Filter the dataframe
filtered_stats_df = stats_df[stats_df['NAME'].isin(selected_states)].reset_index(drop=True)

# Group by NAME (selected states) and compute overall mean and std
overall_mean = filtered_stats_df.drop(columns=['NAME']).mean().round(0).astype(int)
overall_std = filtered_stats_df.drop(columns=['NAME']).std().round(0).astype(int)

# Combine into a single dataframe
overall_stats_df = pd.DataFrame({'Metric': overall_mean.index, 'Mean': overall_mean.values, 'Std': overall_std.values})

# Keep only the first 7 rows
overall_stats_df = overall_stats_df.iloc[:7]

# Display results
overall_stats_df
```
