```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Problem 1: Basic Plotting with Matplotlib

```python
# generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)
# plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, linestyle='-', color='b')
# add title
plt.title("Sin(x) Function")
# add axis labels
plt.xlabel("X")
plt.ylabel("sin(X)")
plt.show()
```

# Problem 2: Exploring Distributions with Seaborn

```python
# load seaborn dataset
tips = sns.load_dataset("tips")

# plot
plt.figure(figsize=(8, 5))
# set attributes
sns.histplot(tips['total_bill'], bins=20, kde=True)
# add title
plt.title("Distribution of Total Bill")
# add labels
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

# Problem 3: Categorical Data Analysis

```python
# create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
# create figures, select data for x and y
sns.boxplot(x="time", y="tip", data=tips, ax=axes[0])
sns.violinplot(x="time", y="tip", data=tips, ax=axes[1])
# add titles
axes[0].set_title("Box Plot of Tips by Time")
axes[1].set_title("Violin Plot of Tips by Time")
plt.show()
```

# Problem 4: Multivariate Analysis with Scatter Plots

```python
# create scatter plot with regression lines
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, aspect=1.5)
# add title
plt.title("Total Bill against Tip by Smoker Status")
plt.show()
```

# Problem 5: Heatmap of Correlation Matrix

```python
# compute correlation for numeric columns
corr_matrix = tips.select_dtypes(include=['number']).corr()
plt.figure(figsize=(8, 6))
# create heatmap, add annotations
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
# add title
plt.title("Correlation Matrix of Tips Dataset")
plt.show()
```

# Problem 6: Pair Plot Analysis

```python
# create pair plot by time and add kdp plots
sns.pairplot(tips, hue="time", diag_kind="kde")
plt.show()
```
