```python
import pandas as pd
```

# import dataset

```python
co2 = pd.read_csv('/content/drive/MyDrive/FWE 458/Lec6/MaunaLoaCO2_monthly_sitename.csv')
co2.head() # display first 5 rows
```

# Filter data

```python
co2 = co2[(co2['Year'] != 1958) & (co2['Year'] != 2022)] # filter out 1958 and 2022
co2 = co2[(co2['numofdays'] >= 0) & (co2['stdofdays'] >= 0)] # ensure numofdays and stdofdays has only non-negative values
```

# Compute statistic

```python
annual_average = co2.groupby("Year")["Monthly_average"].mean().rename("annual_average") # calculate average co2 concentrarion
deseasonalized_annual_average = co2.groupby("Year")["deseasonalized"].mean().rename("deseasonalized_annual_average") # calculate average deseasonalized co2 concentration
std_of_months = co2.groupby("Year")["Monthly_average"].std().rename("std_of_months") # calculate standard deviation

annual_stats = pd.concat([annual_average, deseasonalized_annual_average, std_of_months], axis=1).reset_index() # concatenate columns (axis=1) and keep the original index column (reset_index), convert 'Year' to a column instead of index
```

# Custom function

```python
def co2_classification(annual_average): # define custom function
    if annual_average< 350:  # use if and elif to set thresholds
        return "Low"
    elif 350 <= annual_average <= 400:
        return "Moderate"
    else:
        return "High"
```

```python
annual_stats["CO2_Category"] = annual_stats["annual_average"].apply(co2_classification) # create a new column and apply the classification fucntion on 'annual average'.
```
